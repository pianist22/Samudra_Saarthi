```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('beach_data.csv')
```

```python
columns_to_normalize = ['tavg', 'tmin', 'tmax', 'prcp', 'wdir', 'wspd', 'pres', 
                         'Footfall', 'Wave Height (m)', 'Ocean Currents (m/s)', 
                         'Swell Height (m)', 'Storm Surge', 'Tsunami Alerts', 
                         'Humidity (%)', 'UV Index']
```

```python
scaler = MinMaxScaler()

# Normalize the data
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

# Print the normalized DataFrame
print(df)
```

```python
df.sample(10)
```

```python
df.to_csv('normalized_data.csv', index=False)
```
